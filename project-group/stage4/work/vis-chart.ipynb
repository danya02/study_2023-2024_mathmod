```julia
#data = JSON.parsefile("./solarium/demo5-medium-norepulsion-more/3-fullenergies.json")
data = JSON.parsefile("./solarium/low-velocity-test/6-fullenergies.json")
potential_offset = data["kinetic"][20]
pot = map((x) -> potential_offset - (x/2), data["potential"])[2:5000]
kin = data["kinetic"][2:5000]
offset = (kin[2] + pot[2])/2
plot([pot, kin, kin+pot .- offset])
```
