```julia
using DifferentialEquations, Plots
```

```julia
m01 = 8
m02 = 10
pcr = 50
N = 50
q=1
t1=36
t2=30
p1=10
p2=12
```

```julia
a1 = pcr/(t1^2 * p1^2 * N * q)
a2 = pcr/(t2^2 * p2^2 * N * q)
b = pcr /(t1^2 * p1^2 * t2^2 * p2^2 * N * q)
c1 = (pcr-p1)/(t1 * p1)
c2 = (pcr - p2)/(t2 * p2)
```

```julia
tspan = (0,50)
u0 = [m01, m02];
```

```julia
function eq1(du, u, p, t)
    m1 = u[1]
    m2 = u[2]
    du[1] = m1 - (b/c1)*m1*m2 - (a1/c1)*m1^2
    du[2] = (c2/c1)*m2 - (b/c1)*m1*m2 - (a2/c1)*m2^2
end

function eq2(du, u, p, t)
    m1 = u[1]
    m2 = u[2]
    du[1] = m1 - ((b/c1) + 0.00041)*m1*m2 - (a1/c1)*m1^2
    du[2] = (c2/c1)*m2 - (b/c1)*m1*m2 - (a2/c1)*m2^2
end
```

```julia
prob1 = ODEProblem(eq1, u0, tspan)
sol1 = solve(prob1, AutoTsit5(Rosenbrock23()))
plot(sol1)
```

```julia
prob2 = ODEProblem(eq2, u0, tspan)
sol2 = solve(prob2, AutoTsit5(Rosenbrock23()))
plot(sol2)
```
