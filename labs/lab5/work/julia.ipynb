```julia
using DifferentialEquations, Plots
```

```julia
tspan = (0, 100)
u0 = [7, 21]
```

```julia
function eq(du, u, p, t)
    x=u[1]
    y=u[2]
    du[1] = -0.28x + 0.028x*y
    du[2] = 0.29y - 0.029x*y
end
```

```julia
prob = ODEProblem(eq, u0, tspan, dt=0.01)
sol = solve(prob)
```

```julia
plot(sol)
```

```julia
plot(sol, idxs=(1,2))
```
