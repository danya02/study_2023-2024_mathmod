```julia
using DifferentialEquations, Plots
```

```julia
x0 = [1.7, -0.2]
tspan = (0, 59)
dt = 0.05
```

```julia
function undampened(du, u, p, t)
    x = u[1]
    dx = u[2]
    du[1] = dx
    du[2] = -1.7 * x
end
```

```julia
prob1 = ODEProblem(undampened, x0, tspan, dt=dt)
sol1 = solve(prob1, Tsit5())
```

```julia
plot(sol1, linewidth=2, title="Undampened pendulum")
```

```julia
plot(sol1, vars=(1, 2), title="Undampened phase space")
```

```julia
function dampened(du, u, p, t)
    x = u[1]
    dx = u[2]
    du[1] = dx
    du[2] = -1.7 * x - 1.7 * dx
end
prob2 = ODEProblem(dampened, x0, tspan, dt=dt)
sol2 = solve(prob2, Tsit5())
```

```julia
plot(sol2, linewidth=2, title="Dampened pendulum")
```

```julia
plot(sol2, vars=(1, 2), title="Dampened phase space")
```

```julia
function dampened_energetic(du, u, p, t)
    x = u[1]
    dx = u[2]
    du[1] = dx
    du[2] = -1.7 * x - 1.7 * dx - 0.7cos(2.7t)
end
prob3 = ODEProblem(dampened_energetic, x0, tspan, dt=dt)
sol3 = solve(prob3, Tsit5())
```

```julia
plot(sol3, linewidth=2, title="Dampened pendulum with bonus energy")
```

```julia
plot(sol3, vars=(1, 2), title="Dampened phase space with bonus energy")
```
