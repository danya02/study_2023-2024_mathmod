```julia
using DifferentialEquations, Plots
```

```julia
function model_1(du, u, p, t)
    x = u[1]
    y = u[2]
    dx = -0.441x - 0.773y + sin(2t) + 1
    dy = -0.55x - 0.664y + cos(2t) + 1
    du[1] = dx
    du[2] = dy
end
```

```julia
u0 = [25000, 39000]
tspan = (0.0, 10.0)
```

```julia
condition1(u, t, integrator) = u[1]
cb1 = ContinuousCallback(condition1, SciMLBase.terminate!)
condition2(u, t, integrator) = u[2]
cb2 = ContinuousCallback(condition2, SciMLBase.terminate!)
cb = CallbackSet(cb1, cb2)
```

```julia
prob1 = ODEProblem(model_1, u0, tspan)
sol1 = solve(prob1, Tsit5(), callback=cb, abstol = 1e-15, dt = 0.0001)
```

```julia
plot(sol1, linewidth = 2, xaxis = "t", yaxis = "count", label = ["X" "Y"])
```

```julia
function model_2(du, u, p, t)
    x = u[1]
    y = u[2]
    dx = -0.399x - 0.688y + sin(2t) + 2
    dy = -0.299x*y - 0.811y + cos(3t) + 1
    du[1] = dx
    du[2] = dy
end
```

```julia
prob2 = ODEProblem(model_2, u0, tspan)
sol2 = solve(prob2, Tsit5(), callback=cb, abstol = 1e-15, dt = 0.0001)
```

```julia
plot(sol2, linewidth = 2, xaxis = "t", yaxis = "count", label = ["X" "Y"])
```
