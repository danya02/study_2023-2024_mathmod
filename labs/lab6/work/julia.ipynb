```julia
using DifferentialEquations, Plots
```

```julia
tspan = (0, 50)
u0 = [8124-124-30, 124, 30];
alpha = 0.1;
beta = 0.2;
```

```julia
function eq_low(du, u, p, t)
    s=u[1]
    i=u[2]
    r=u[3]
    du[1] = 0
    du[2] = -beta * i
    du[3] = beta * i
end

function eq_high(du, u, p, t)
    s=u[1]
    i=u[2]
    r=u[3]
    du[1] = -alpha * s
    du[2] = alpha * s - beta * i
    du[3] = beta * i
end
```

```julia
prob1 = ODEProblem(eq_low, u0, tspan, dt=0.01)
sol1 = solve(prob1)
prob2 = ODEProblem(eq_high, u0, tspan, dt=0.01)
sol2 = solve(prob2);
```

```julia
plot(sol1)
```

```julia
plot(sol2)
```
