```julia
using DifferentialEquations, Plots
```

```julia
tspan = (0, 30)
u0 = [12];
global N
N = 1420
```

```julia
function eq1(du, u, p, t)
    N=1420
    n = u[1]
    du[1] = (0.7 + 0.000012*n)*(N - n)
end

function eq2(du, u, p, t)
    N=1420
    n = u[1]
    old_dn = du[1]
    du[1] = (0.00003 + 0.5*n)*(N - n)
    if old_dn > du[1]
        println("Inflection: t=", t)
    end
end

function eq3(du, u, p, t)
    N=1420
    n = u[1]
    du[1] = (0.57*sin(t) + 0.38*cos(13*t) * n)*(N - n)
end
```

```julia
prob1 = ODEProblem(eq1, u0, tspan, dt=0.01)
sol1 = solve(prob1, AutoTsit5(Rosenbrock23()))
plot(sol1)
```

```julia
prob2 = ODEProblem(eq2, u0, (0,3), dt=0.01)
sol2 = solve(prob2, AutoTsit5(Rosenbrock23()));
plot(sol2)
```

```julia
prob3 = ODEProblem(eq3, u0, (0,2), dt=0.01)
sol3 = solve(prob3, AutoTsit5(Rosenbrock23()));
plot(sol3)
```
